```python
import dash
import dash_core_components as dcc
import plotly.express as px
import pandas as pd
insurance = pd.read_csv('insurance.csv')
def dynamic_layout():
    
    return dcc.Graph(figure=px.bar(insurance, x=insurance.region, y=insurance.age,
              hover_data=['sex', 'smoker'], color='sex')
                            .update_yaxes(tickprefix="ages")
                            .update_layout(legend_orientation="sex"))
app = dash.Dash(__name__)
app.layout = dynamic_layout
app.run_server()
```
